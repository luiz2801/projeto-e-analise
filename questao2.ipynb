### ExecuûÏûÈo do Algoritmo de Floyd-Warshall

| Vûˋrtice               | **1** | **2** | **$\beta_1$** | **$\beta_2$** | **$\beta_3$** |

| **$\alpha_2$** | 2 | 3 | 4 | 6 | 5 |
| **$\alpha_3$** | $\infty$ | 1 | 4 | 6 | 3 |

Com a matriz de adjacûˆncias posta ûˋ possûÙvel aplicar a tûˋcnica de programaûÏûÈo linear para solucionar o problema.
