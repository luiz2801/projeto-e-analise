# Passo 1: Formulação como Problema de Caminho Máximo (Grafo Dirigido)

Para modelar o planejamento do projeto como um **Problema de Caminho Máximo**, representamos o cronograma por meio de uma rede de atividades (Grafo Direcionado Acíclico - DAG):

* **Vértices (Nós):** Representam os marcos/eventos do projeto (instantes de tempo de início ou término) ou as próprias atividades. Na modelagem tradicional de redes (Activity-on-Node), cada tarefa é um nó. Para fecharmos o grafo perfeitamente, adicionamos um nó fictício de **Início** e um de **Fim**.
* **Arestas (Arcos direcionados):** Representam a precedência temporal. Se a tarefa $A$ é pré-requisito da tarefa $B$, existe um arco $A \rightarrow B$.
* **Pesos:** O peso associado ao nó (ou ao arco que sai dele) é a **duração da tarefa** em dias.

### Por que um Problema de Caminho Máximo?
O tempo total do projeto é determinado pelo caminho de maior dependência sequencial do início ao fim (o caminho mais "longo" em termos de tempo). Encontrar o **tempo mínimo** para terminar a obra equivale matematicamente a encontrar o **Caminho Máximo** neste grafo direcionado, pois o projeto só termina quando a última sequência de atividades dependentes for concluída.

# O Algoritmo do Caminho Crítico (CPM / PERT)

Para resolver o problema de caminho máximo e determinar os prazos de cada atividade, aplicamos o método dos passos adiante (**Ida**) e para trás (**Volta**).

### 1. Passo de Ida (Tempos Mais Cedo)
Determina o momento mais cedo em que uma tarefa pode começar ($ES$ - *Early Start*) e terminar ($EF$ - *Early Finish*).
* **Início Mais Cedo ($ES$):** Se a tarefa não tem pré-requisitos, $ES = 0$. Se depende de outras tarefas, $ES$ é o **máximo** entre os tempos de término mais cedo ($EF$) de todos os seus pré-requisitos imediatos:
    $$ES_i = \max \, \{ EF_j \mid j \text{ é pré-requisito de } i \}$$
* **Término Mais Cedo ($EF$):** $$EF_i = ES_i + \text{Duração}_i$$

*O maior $EF$ obtido na última tarefa determina a **Duração Mínima do Projeto**.*

### 2. Passo de Volta (Tempos Mais Tarde)
Determina o limite máximo de tempo em que uma tarefa pode começar ($LS$ - *Late Start*) e terminar ($LF$ - *Late Finish*) sem atrasar o projeto.
* Fixamos o $LF$ da última tarefa igual à duração total do projeto calculada no passo de ida.
* **Término Mais Tarde ($LF$):** É o **mínimo** entre os tempos de início mais tarde ($LS$) de todas as tarefas que dependem dela:
    $$LF_i = \min \, \{ LS_k \mid i \text{ é pré-requisito de } k \}$$
* **Início Mais Tarde ($LS$):**
    $$LS_i = LF_i - \text{Duração}_i$$

### 3. Cálculo de Folgas
A **Folga Total ($FT$)** indica quantos dias uma tarefa pode atrasar sem impactar o prazo final da obra:
$$FT_i = LF_i - EF_i = LS_i - ES_i$$

# Conclusões 

Com base na planilha de cálculo obtida pelo método do Caminho Crítico:

* **Tempo Mínimo de Execução:** O número mínimo de dias necessário para terminar o conjunto completo de tarefas é de **36 dias** (calculado pelo término mais cedo da tarefa 9).
* **Tarefas Críticas (Folga Nula):** As tarefas que não toleram nenhum tipo de atraso são as tarefas **1, 2, 3, 4, 6, 8 e 9**. A sequência delas forma o **Caminho Crítico** do projeto:
    $$\text{Início} \rightarrow 1 \rightarrow 2 \rightarrow 3 \rightarrow 4 \rightarrow 6 \rightarrow 8 \rightarrow 9 \rightarrow \text{Fim}$$
* **Tarefas com Folga:** Duas tarefas possuem flexibilidade no cronograma:
    1.  **Tarefa 5 (Colocação das tábuas externas):** Possui uma folga de **2 dias**. Ela pode começar no dia 14 (mais cedo) ou até o dia 16 (mais tarde).
    2.  **Tarefa 7 (Colocação das janelas e portas):** Possui uma folga de **12 dias**. Ela pode iniciar no dia 14 (mais cedo) ou esperar até o dia 26 (mais tarde) para começar sem comprometer a pintura (Tarefa 9).
